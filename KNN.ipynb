```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from IPython import display
```

> It will first calculate the distance between the data points

> find the K nearest data points 

> The majority of votes will be the prediction

```python
display.Image("photo_2021-01-04_10-50-58.jpg")
```

```python
X, y = make_moons(noise = 0.3)

data = pd.DataFrame(X, columns=["X1", "X2"])
data['y'] = y


plt.title('Plot 1')
sns.scatterplot(data=data, x= "X1", y= "X2", hue="y", style="y", s=200)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model = KNeighborsClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```
